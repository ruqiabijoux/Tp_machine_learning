```python
#repartition des scores
sns.countplot(x='math score', data=mbb)
plt.title("R√©partition de score en math")
plt.show()
```

```python
#repartition des scores
sns.countplot(x='writing score', data=mbb)
plt.title("R√©partition de score en R√©daction")
plt.show()
```

```python
#repartition des genres
sns.countplot(x='gender', data=mbb)
plt.title("R√©partition des genres")
plt.show()
```

```python
# scores selon le genre
sns.histplot(data=mbb, x='math score', hue='gender', kde=True, bins=30)
plt.title("Score selon le genre")
plt.show()
```

```python
#genre par score en lecture et en math
sns.histplot(x='reading score', y='gender', hue='math score', data=mbb)
plt.title("genre par score en lecture et en math")
plt.show()
```

```python
#matrice de corr√©lation
plt.figure(figsize=(10,8))
sns.heatmap(mbb.corr(), annot=True, cmap='coolwarm')
plt.title("matrice de corr√©lation")
plt.show()
```
